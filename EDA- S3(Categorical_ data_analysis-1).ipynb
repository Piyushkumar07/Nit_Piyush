**Bar-Chart**
- in order to draw bar chart  
- we required i catogrical column 
- one numerical column
- dataframe :continent_value_count_DataFrame

```python
#plt.bar(<)
plt.figure(figsize=(10,6))
plt.bar('continent','count',data=continent_value_count_DataFrame)
plt.xlabel("continent")# x-axis name
plt.ylabel('count')# y-axis name
plt.title("Bar-Chart")# title of the chart
plt.savefig('Continent_bar.jpg')
plt.show()
```

$Count-Plot$
- can use by seaborn package 
- it required only **entire dataframe** and **catogrical column**
- entire dataframe name=df
- catogricaalcolumn name=continent

```python
plt.figure(figsize=(20,12))
l=df['continent'].value_counts().keys()# order provide automatically
sns.countplot(data=df,x='continent',order=l)
plt.xlabel("continent")# x-axis name
plt.ylabel('count')# y-axis name
plt.title("Bar-Chart")# title of the chart
plt.savefig('Continent_bar1.jpg')
plt.show()
```
