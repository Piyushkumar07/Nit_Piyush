```python
path=r"C:\Users\piyus\OneDrive\Data_Files\visadataset.csv"
visa_df=pd.read_csv(path)
visa_df
```

```python
d1=visa_df["prevailing_wage"]
d2=len(d1)
Q1=np.percentile(visa_df["prevailing_wage"],25)
Q2=np.percentile(visa_df["prevailing_wage"],50)
Q3=np.percentile(visa_df["prevailing_wage"],75)
lb=Q1-1.5*d2
ub=Q3+1.5*d2
l1=[]
for i in d1:
    if (i<lb) or (i>ub):
        l1.append(i)
    else:
        l1.append(i)
visa_df["prevailing_wage_new"]=l1
```

```python
d1=visa_df["prevailing_wage"]
d2=len(d1)
Q1=np.percentile(visa_df["prevailing_wage"],25)
Q2=np.percentile(visa_df["prevailing_wage"],50)
Q3=np.percentile(visa_df["prevailing_wage"],75)
lb=Q1-1.5*d2
ub=Q3+1.5*d2
c1=visa_df["prevailing_wage"]<lb
c2=visa_df["prevailing_wage"]>ub
c=c1|c2
median=visa_df["prevailing_wage"].median()
t=median
f=visa_df["prevailing_wage"]
visa_df["prevailing_wage"]=np.where(c,t,f)
plt.boxplot(visa_df["prevailing_wage"])
plt.show()
```
