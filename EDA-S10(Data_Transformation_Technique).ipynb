**Data Transformation Technique** 
- Generally used for to Convert Normal distribution 
- Because all statistical math analysis by assumption Data follows Normal
  distribution 
- it is also avoid skew ness also 
- We have some important transformation 
     - Log transformation 
     - Exponential Transformation 
     - Reciprocal Transformation
     - Square root transformation 
     - Power transformation


***Step-1***
- Read the Required package

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

****Step-2****
   
   
   ****Exponential- data****

```python
exp_data=np.random.exponential(size=10000)
exp_data[:10]
# we are consider exponential data set with random 1k obsarvation 
# using numpy package
```

```python
plt.hist(exp_data,bins=40,label="Exponential")
plt.legend()
plt.show()
```

**Norm_data**

```python
Norm_data=np.random.normal(size=1000)
plt.hist(Norm_data)
```

**Step-3**
$Log$ $Transformation$
- np.log is used for long transformation 
- Generally log transformation will not convert data into normal 
- it avoids skewness
- np.log means natural algorithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data[:10]
```

```python
plt.hist(log_data,bins=40,label='Log Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1).hist(exp_data,bins=40,label="Exponential")
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=40,label="Log Transformation")
plt.legend()
plt.show()
```

**Square root Transformation**

```python
print(25**2)
print(25**(1/2))
print(np.sqrt(25))
```

```python
sqrt_data=np.sqrt(exp_data)
sqrt_data
```

$Step-6$
**Power Transformation**
- It is related to sklearn package
- Package name: sklearn.preprocessing
- Method name: Power Transformer
- Inside Box-Cox , yeo-jhonson

```python
exp_data
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
pt_data=pt.fit([exp_data]).transform([exp_data])
plt.hist(pt_data[0])
plt.show()
```
