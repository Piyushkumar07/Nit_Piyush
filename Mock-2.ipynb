```python
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            return year % 400 == 0
        else:
            return True
    else:
        return False

count = 0
leap_years = []

for year in range(1700, 2025):
    if is_leap_year(year):
        count += 1
        leap_years.append(year)

print("Leap years between 1700 and 2025:", leap_years)
print("Total number of leap years:", count)
```

```python
#10
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
num_terms = int(input("Enter the number of terms: "))
fibonacci(num_terms)
```

```python
#5
data = {'Category': ['A', 'B', 'C', 'D', 'E'], 'Value': [10, 25, 18, 32, 15]}
df = pd.DataFrame(data)
# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(df['Category'], df['Value'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot of Data')  
plt.show()

# Histogram plot
plt.figure(figsize=(8, 6))
plt.hist(df['Value'], bins=5, edgecolor='black') 
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.show()

# Heatmap 
heatmap_data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
plt.figure(figsize=(8, 6))
plt.pcolor(heatmap_data, cmap='hot') 
plt.colorbar(label='Value')
plt.xlabel('Column Index')
plt.ylabel('Row Index')
plt.title('Heatmap of Data')
plt.show()

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Category'], df['Value'], color='blue', alpha=0.7) 
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Scatter Plot of Data')
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%', startangle=140) 
plt.title('Pie Chart of Data')
plt.show()
```
